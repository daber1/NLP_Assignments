- Each smiley is a separate token
- Each hashtag is an individual token. Each user reference is an individual token

## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import Counter, defaultdict
import re
from operator import itemgetter

class BPETokenizer:
    def __init__(self, vocab_size: int) -> None:
        """
        Initializes the Tokenizer with the specified vocabulary size.

        Args:
          vocab_size : int
            The size of the vocabulary
        """
        self.vocab_size = vocab_size
        self.vocab = defaultdict(int)
        self.merges = {}

    def train(self, corpus: List[str]) -> None:
        """
        Trains the Tokenizer

        Args:
          corpus : list[str]
            The corpus of texts
        """
        # Count occurrences of symbols
        for text in corpus:
            words = text.split()
            for word in words:
                token = ' '.join(list(word)) + ' </w>'
                self.vocab[token] += 1

        # Perform merges iteratively
        for _ in range(self.vocab_size):
            pairs = defaultdict(int)
            for word, freq in self.vocab.items():
                symbols = word.split()
                for i in range(len(symbols)-1):
                    pairs[symbols[i],symbols[i+1]] += freq
            best = max(pairs, key=pairs.get)
            merged_vocab = {}
            bigram = re.escape(' '.join(best))
            p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
            for word in self.vocab:
                w_out = p.sub(''.join(best), word)
                merged_vocab[w_out] = self.vocab[word]
            self.merges[best] = pairs[best]
            self.vocab = merged_vocab.copy()

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenizes the input text using the learned vocabulary

        Args:
          text : str
            The input text to be tokenized

        Returns:
          list: A list of tokens obtained from the input text
        """
        words = text.split()
        answer = []
        for word in words:
            answer.extend(self.split_word(word))
        return answer

    def split_word(self, word: str) -> List[str]:
        """
        Splits a word into tokens using the learned vocabulary

        Args:
          word : str
            The word to be split

        Returns:
          list: A list of tokens obtained from the input word
        """
        if word in self.vocab:
            return [word]

        # Initialize a list to store tokens
        tokens = []

        # Start from the longest possible token length
        for l in range(len(word), 0, -1):
            for i in range(len(word) - l + 1):
                subword = word[i:i+l]
                if subword + '</w>' in self.vocab:
                    tokens.append(subword)
                    tokens.extend(self.split_word(word[i+l:]))
                    return tokens
        tokens.extend([char for char in word])
        return tokens
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
# Dataset
data = []
for i in range(1,6):
  with open(f'file{i}', 'r') as f:
    tweets = f.readlines()
    for tweet in tweets:
      data.append(tweet.strip())
```

```python
# Define the vocab_sizes to try
vocab_sizes = [100, 250, 500, 750, 1000]
```

```python
# Initialize lists to store average token lengths for each vocab_size
avg_token_lengths = []

# Iterate over each vocab_size
for vocab_size in vocab_sizes:
    # Initialize and train the tokenizer
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(data)

    # Tokenize tweets and calculate average token length
    total_tokens = 0
    total_tweets = len(data)
    for tweet in data:
        tokens = tokenizer.tokenize(tweet)
        total_tokens += len(tokens)

    # Calculate average token length for the current vocab_size
    avg_token_length = total_tokens / total_tweets
    avg_token_lengths.append(avg_token_length)


import matplotlib.pyplot as plt
plt.plot(vocab_sizes, avg_token_lengths, marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Token Length')
plt.title('Dependency of Average Token Length on Vocabulary Size')
plt.grid(True)
plt.show()
```

On average, increasing the vocabulary size decreses the average token length of tweets.
